```python
import pandas as pd
import glob
import numpy as np
```

```python
path = "C:/Users/gianl/Desktop/Gi/Supsi/MartaLauraGiada/Laura/Tweets/"
entries = glob.glob(path + "tweets.parquet")[0]
df = pd.read_parquet(entries, engine="fastparquet")
```

```python
df.head()
```

```python
df_urls = df.loc[df['urls'] != '[]']
val = [x.split()[3].translate({ord("'"): None}).replace(",","") for x in df_urls["urls"]]
val = [x if ("https://youtu.be" in x) or ("https://www.youtube.com" in x) else np.nan for x in val]
```

```python
x = pd.Series(val).dropna()
x = x.value_counts()
f = list(x.keys())
c = list(x.values)
```

```python
pd.DataFrame(list(zip(f,c)), columns=["URL", "VAL"]).to_csv("C:/Users/gianl/Desktop/Gi/Supsi/BachelorProject/youtube_checker/data/file_giada.csv", 
                                       line_terminator="\n", index=False, encoding="utf-8")
```
