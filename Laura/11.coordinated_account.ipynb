```python
import pandas as pd
from tqdm.notebook import tqdm
```

```python
path_df = "C:/Users/gianl/Desktop/Gi/Supsi/Vaccines_Discussion_Italy/Laura/Tweets/"
path = "C:/Users/gianl/Desktop/Gi/Supsi/Vaccines_Discussion_Italy/Giada/Files/"
```

```python
good = pd.read_csv(path + "good.csv")
bad = pd.read_csv(path + "bad.csv")
good_lst = list(good["screen_name"])
bad_lst = list(bad["screen_name"])
good_rt = pd.read_csv(path_df + "good_rt.csv")
bad_rt = pd.read_csv(path_df + "bad_rt.csv")
good_lst = list(good["screen_name"])
bad_lst = list(bad["screen_name"])
good_rt_lst = list(good_rt["screen_name"])
bad_rt_lst = list(bad_rt["screen_name"])
```

```python
print(f"Numero di utenti {len(good_rt_lst)}")
print(f"Numero di utenti {len(bad_rt_lst)}")
```

```python
retweets = pd.read_parquet(path_df + "retweets.parquet", engine = "fastparquet")
retweets.head()
```

```python
retweets['creato'] = pd.to_datetime(retweets['created_at'], format='%a %b %d %H:%M:%S %z %Y')
retweets['rt_creato'] = pd.to_datetime(retweets['rt_created_at'], format='%a %b %d %H:%M:%S %z %Y')
```

```python
retweets['seconds_retweet'] = (retweets['creato']-retweets['rt_creato']).astype('timedelta64[s]')
```

```python
seconds = 60
consecutive_retweets = retweets[retweets['seconds_retweet']<=seconds]
```

```python
consecutive_retweets[['id','user_screen_name','rt_user_screen_name']]
```

```python
lst = []
for tweet in tqdm(consecutive_retweets.itertuples()):
    value = tweet.user_screen_name+"-"+tweet.rt_user_screen_name
    if tweet.user_screen_name != tweet.rt_user_screen_name:
#         print(value)
        lst.append(value)
```

```python
serie = pd.Series(lst).value_counts()
for index,value in serie.iteritems():
    if value > 1:
        print(f"{index} - {value}")
```

```python
x = list(pd.Series(consecutive_retweets["user_screen_name"]).value_counts().sort_values(ascending=False).keys())
```

```python
goodd = 0
badd = 0
gooddd = 0
baddd = 0
lst = []
lst_quantity = []
for i in tqdm(consecutive_retweets["user_screen_name"]):
    if (i in good_rt_lst) | (i in good_lst):
        gooddd = gooddd + 1
    elif (i in bad_rt_lst) | (i in bad_lst):
        baddd = baddd + 1
        lst_quantity.append(i)
for i in tqdm(x):
    if (i in good_rt_lst) | (i in good_lst):
        goodd = goodd + 1
    elif (i in bad_rt_lst) | (i in bad_lst):
        badd = badd + 1
        lst.append(i)
print(f"Good users: {goodd}")
print(f"Bad users: {badd}")
print(f"Good q.ty: {gooddd}")
print(f"Bad q.ty: {baddd}")
```
