# Suspended

```python
tweets = pd.read_csv(path_df + r"\tweets.csv", lineterminator="\n", low_memory=False, encoding="utf-8")
```

```python
suspended_files = glob.glob(r"..\script_directory_output\suspended_users\out*.csv")
df = pd.DataFrame()
for file in suspended_files:
    temp = pd.read_csv(file, lineterminator="\n", encoding="utf-8", low_memory=False)
    df = df.append(temp)
df.drop_duplicates(subset=["name"], inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
lst_suspended = df[df["status"].str.contains("'code': 63")]
lst_suspended
```

```python
tweets = tweets[tweets["user_screen_name"].isin(list(lst_suspended["name"]))]
tweets.head()
```

```python
splitted = utils.split(tweets)
labels = list(splitted.keys())
values = list(splitted.values())
utils.print_pie_chart4("Suspended split", labels, values)
```
