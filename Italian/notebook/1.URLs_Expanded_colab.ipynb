```python
### Change Path & Name of Files
path_url = 'C:/Users/gianl/Desktop/Gi/Supsi/MartaLauraGiada/Laura/Files/'
```

```python
compressed_url_df = pd.read_csv(path_url+'urls_processed.csv', lineterminator='\n', low_memory=False)
```

```python
compressed_url_df
```

```python
compressed_url_df.drop_duplicates(inplace=True)
```

```python
compressed_url_df = compressed_url_df.rename(columns={"url": "COMPRESSED"})
compressed_url_df = compressed_url_df.rename(columns={"expanded": "EXPANDED"})
```

```python
compressed_url_df
```

```python
compressed_url_df = compressed_url_df.dropna(subset=['EXPANDED'])
```

```python
compressed_url_df
```

```python
from tqdm.notebook import tqdm
def substitute_compressed_url(url):
    indices = [i for i, x in enumerate(url.split()) if 'expanded_url' in x]
    for idx in range(len(indices)):
        expanded_url = url.split()[int(indices[idx]+1)].replace("'", "").replace(",", "")
        if expanded_url in compressed_url_df['COMPRESSED'].values:
            url = url.replace(expanded_url, compressed_url_df[compressed_url_df['COMPRESSED']==expanded_url]['EXPANDED'].values[0])
    return url
```

```python
data_path = 'C:/Users/gianl/Desktop/Gi/Supsi/MartaLauraGiada/Laura/Tweets/'
```

```python
tweets = pd.read_parquet(data_path+'tweets.parquet', engine="fastparquet")
```

```python
tweets.head()
```

```python
tweets_with_urls = tweets.loc[tweets['urls'] != '[]']
```

```python
lst = []
for i in tqdm(tweets_with_urls["urls"]):
    lst.append(substitute_compressed_url(i))
tweets_with_urls["urls"] = lst
```

```python
destination_path = "C:/Users/gianl/Desktop/Gi/Supsi/MartaLauraGiada/Laura/Tweets/"
```

```python
tweets_with_urls.to_parquet(destination_path+"tweets_w_urls_decompressed.parquet", index=False)
```

```python
def get_list_by_keyword(subject, keyword):
    list_by_key = []
    indices = [i for i, x in enumerate(subject.split()) if keyword in x]
    if len(indices) == 0:
        return None
    list_by_key = [None] * len(indices)
    for idx in range(len(indices)):
        list_by_key[idx] = subject.split()[int(indices[idx]+1)].replace("'", "").replace(",", "")
    return list_by_key
```

```python
#extract list of urls from tweets
tweets_with_urls['expanded_url'] = tweets_with_urls.apply(lambda x: get_list_by_keyword(x.urls, 'expanded_url'), axis=1)
print('There are tweets_with_urls without any url: ',tweets_with_urls['expanded_url'].isnull().values.any()) #check if all tweets with urls have expanded_url. If False it's OK
```

```python
tweets_with_urls.head()
```

```python
#create df with one row for every url
url_df = tweets_with_urls['expanded_url'].apply(pd.Series).reset_index().melt(id_vars='index').dropna()[['index', 'value']].set_index('index')
```

```python
#Find number of urls in every tweet
tweets_with_urls['number_of_urls'] = tweets_with_urls.apply(lambda x: len(x.expanded_url), axis=1)
```

```python
tweets_with_urls.parquet(destination_path+"tweets_w_urls_not_compressed_and_expanded.parquet", index=False)
```

```python
url_df.to_csv(destination_path+"url_df.csv", index=False)
```
